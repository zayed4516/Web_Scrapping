```python
l=[]

for i in range(len(all_jobs)):
    job_name = all_jobs[i].find('h2').text.strip()

    company = [a.get_text(strip=True) for a in all_jobs[i].find_all('a', {'class': 'css-ipsyv7'})]
    job_place = [span.get_text(strip=True) for span in all_jobs[i].find_all('span', {'class': 'css-16x61xq'})]

    company = ''.join(company)
    job_place = ', '.join(job_place)

    # posted date (handling None safely)
    posted_div = all_jobs[i].find('div', {'class': 'css-eg55jf'})
    posted_date = posted_div.get_text(strip=True) if posted_div else None

    job_types_div = all_jobs[i].find('div', class_='css-5jhz9n')
    job_types = [span.get_text(strip=True) for span in job_types_div.find_all('span')] if job_types_div else []

    categories = [a.get_text(strip=True) for a in all_jobs[i].find_all('a', class_='css-o171kl')]
    categories += [a.get_text(strip=True) for a in all_jobs[i].find_all('a', class_='css-5x9pm1')]
    l.append({
        'job_name':job_name,
        'company':company,
        'job_place':job_place,
        'job_types':" / ".join(job_types),
        'categories':" · ".join(categories),
        'posted_date':posted_date
    })
      
    print(job_name)                         
    print(f"{company} - {job_place}")        
    print(posted_date)                      
    print(" / ".join(job_types))            
    print(" · ".join(categories))           
    print("-" * 60)  
```

```python
import pandas as pd
df=pd.DataFrame(l)
df
```
